```python
import tensorflow as tf
import numpy as np
import cv2
import os
import PIL
import random
import matplotlib.pylab as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def plotImages(images_arr):
    fig, axes=plt.subplots(1,10,figsize=(20,20))
    axes=axes.flatten()
    for img,ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis("off")
    plt.tight_layout()
    plt.show()
    
```

```python
gen= ImageDataGenerator(rotation_range=10,width_shift_range=0.1,
    height_shift_range=0.1,shear_range=0.15,zoom_range=0.1,channel_shift_range=10.,horizontal_flip=True)
```

```python
chosen_image=random.choice(os.listdir(r'C:\Users\nanom\OneDrive\Desktop\zircon thin section'))
chosen_image
```

```python
image_path = r'C:\Users\nanom\OneDrive\Desktop\zircon thin section' + '\\' + chosen_image
assert os.path.isfile(image_path)
```

```python
image = np.expand_dims(plt.imread(image_path), axis=0)
plt.imshow(image[0])
```

```python
aug_iter=gen.flow(image)
aug_images=[next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotImages(aug_images)
output_directory = r'C:\Users\nanom\OneDrive\Desktop\zircon augmented'


for i, augmented_image in enumerate(aug_images):
    image_path = os.path.join(output_directory, f'augmented_image_{i + 1}.png')
    print(f"Saving {image_path}")
    plt.imsave(image_path, augmented_image)
```
